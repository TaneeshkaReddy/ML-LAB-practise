# A*

```python
import heapq

def astar(graph, start, goal, heuristic):
    pq=[]
    heapq.heappush(pq,(0+heuristic[start],start))
    visited=set()
    parent={start:None}
    g_cost={start:0}
    
    while pq:
        currh,currn=heapq.heappop(pq)
        if currn in visited:
            continue
        else:
            visited.add(currn)
            
        if visited == goal:
            break
            
        for neigh in graph[currn]:
            new_cost=g_cost[currn]+cost[(currn,neigh)]
            if neigh not in g_cost or new_cost<g_cost[neigh]:
                g_cost[neigh]=new_cost
                f_cost=new_cost+heuristic[neigh]
                heapq.heappush(pq,(f_cost,neigh))
                parent[neigh]=currn
                
    path=[]
    node=goal
    while node is not None:
        path.append(node)
        node=parent[node]
    path.reverse()
        
    return path,g_cost.get(goal,0)


graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F', 'G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Example heuristic values (assumed for demonstration)
heuristic = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 0,
    'E': 2,
    'F': 3,
    'G': 1
}

cost = {
    ('A', 'B'): 1,
    ('A', 'C'): 1,
    ('B', 'D'): 1,
    ('B', 'E'): 3,
    ('C', 'F'): 5,
    ('C', 'G'): 2
}

start = 'A'
goal = 'D'

path,cost = astar(graph, start, goal, heuristic)
print("A* PATH is ", path," and cost value is ",cost)
  
```
