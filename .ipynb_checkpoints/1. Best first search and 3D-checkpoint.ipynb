```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('ToyotaCorolla.csv')
x=df['KM']
y=df['Doors']
z=df['Price']

ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='jet')
ax.set_title("3D surface plot")
plt.show()
```

# BEST FIRST SEARCH

```python
import heapq

def bfs(graph, start, goal, heuristic):
    pq=[]
    heapq.heappush(pq,(heuristic[start],start))
    visited=set()
    parent={start:None}
    cost={start:0}
    
    while pq:
        currh,currn=heapq.heappop(pq)
        if currn in visited:
            continue
        else:
            visited.add(currn)
            
        if visited == goal:
            break
            
        for neigh,edgecost in graph[currn]:
            if neigh not in visited:
                heapq.heappush(pq,(heuristic[neigh],neigh))
                parent[neigh]=currn
                cost[neigh]=cost[currn]+edgecost
                
    path=[]
    node=goal
    while node is not None:
        path.append(node)
        node=parent[node]
    path.reverse()
        
    return path,cost.get(goal,0)


graph = {
    'A': [('B', 1), ('C', 2)],
    'B': [('D', 3), ('E', 4)],
    'C': [('F', 5), ('G', 6)],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

# Example heuristic values (assumed for demonstration)
heuristic = {
    'A': 6,
    'B': 4,
    'C': 4,
    'D': 0,
    'E': 2,
    'F': 3,
    'G': 1
}

start = 'A'
goal = 'D'

path,cost = bfs(graph, start, goal, heuristic)
print("Best First Search Path is ", path," and cost value is ",cost)
    
    
```
