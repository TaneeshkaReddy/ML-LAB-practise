```python

import pandas as pd
import matplotlib.pyplot as plt


df=pd.read_csv('ToyotaCorolla.csv')
x=df['KM']
y=df['Doors']
z=df['Price']

plt.boxplot([x,y,z])

plt.xticks([1,2,3],["Price","HP","KM"])

plt.show()
```
